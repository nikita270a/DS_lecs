#### Визуализация

```python
import matplotlib.pyplot as plt
```

```python
data.corr()['Hormonal Contraceptives'].abs().sort_values(ascending=False).head(15).plot(kind='bar', figsize=(10,5))
plt.title('Correlcation with Hormonal Contraceptives feature (top 15)')
```

Можно строить графики по осям (удобно для нескольких графиков сразу)

```python
x = np.linspace(1, 10, 20)
fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, x ** 2, 'r')
axes.plot(x, x ** 3, 'b*--')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.legend([r'$x^2$', 'x^3'], loc = 0)
```

Можно настроить оформление, поменяв детали

```python
fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, x ** 2, 'r^-', label = r'$y = x^2$', markersize=8, markerfacecolor="yellow", 
          markeredgewidth=1, markeredgecolor="green")
axes.plot(x, x ** 3, 'b*--', label = r'$y = x^3$', alpha = 0.5)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.legend(loc = 0, fontsize = 18)
```

Можно накладывать графики друг на друга для наглядности

```python
fig = plt.figure()
axes = fig.add_axes([0.0, 0.0, 1.0, 1.0])
bins = 20
index = np.arange(bins)
axes.hist(data[data['Smokes'] == 1]['Age'].dropna(), bins=bins, alpha=0.6, label='smoking', normed=True)
axes.hist(data[data['Smokes'] == 0]['Age'].dropna(), bins=bins, alpha=0.6, label='non smoking', normed=True)

axes.legend()
axes.set_xlabel('Age', fontsize=18)
axes.set_ylabel('Count', fontsize=18)
axes.set_title('Age by smoking', fontsize=18)
```

## Что такое машинное обучение?

"Машинное обучение - это полуавтомтизированное извлечение знаний из данных"

- **знаний из даты**: что мы можем вынести, имею исходные данные?
- **автоматическое извлечение**: компьютер самостоятельно строит закономерности
- **полуавтоматическое**: человек нужен для принятия умных решений

### 2 основных типа машинного обучения: 

#### С учителем
Пример: является ли данное письм оспамом? 

Есть ответ, который мы предсказываем

![scikit-learn logo](pics/01_spam_filter.png)

#### Обучение с учителем  (есть четкие ответы для обучающих данных)

- Задача регрессии — задача с вещественной целевой переменной, где основная цель предсказать значение близкое к фактическому в смысле некоторого функционала ошибки. Например, предсказать зарплату человека по его опыту, образованию, месту жительства.
![scikit-learn logo](pics/regression.jpeg)

- Задача классификации (в том числе многоклассовая и многоклассовая с пересекающимися классами) — задача, где целевая переменная принимает значение из некоторого фиксированного конечного множества значений (или несколько в случае пересекающихся классов). Например, задача определения типа автомобиля: легковой, грузовой, пассажирский.
![scikit-learn logo](pics/calssification.png)


- Задача ранжирования — задача, в которой необходимо отсортировать объекты выборки по некоторому критерию (релеватности). При этом иногда нужно сортировать все объекты глобально, а иногда только внутри некоторых групп (например, в поиске необходимо сортировать ссылки для каждого запроса отдельно). Типичный пример — поисковая система.
![scikit-learn logo](pics/filtering.jpeg)

#### Обучение без учителя (нет четких ответов для обучающих данных)

- Задача кластеризации — задача, в которой выборка не имеет какой-то целевой переменной, а необходимо разделить объекты на группым по некоторым свойствам. На практике нередко сложно оценить качество при решении этой задачи. Пример — разделить пользователей сервиса на группы для дальнейшего исследования клиентов. 
![scikit-learn logo](pics/clustering.png)


- Задача поиска выбросов — задача, похожая на бинарную классификацию с преобладанием объектов одного класса с тем отличием, что объекты "редкого" класса могут быть не похожими между собой, а лишь отличаться от объектов основного класса. Кроме этого, на практике объектов-выбросов в выборке может и не быть. Например, задача поиска фальшивых монет. Если быть строже, задачи этого класса можно разделить на детектирование выбросов и новизны.
![scikit-learn logo](pics/anomaly.png)



- Задача обучения с подкреплением — задача, в которой обучающая система учится воздействовать со средой, не имея сведений о ней. Например, двуногий робот учится управлять своими шарнирами, двигаясь по ровной поверхности. Такие задачи достаточно специфичны и не будут рассматриваться в этом курсе.
![scikit-learn logo](pics/RL.png)


- Задача сжатия размерности — задача, в которой система, имея большое количество входных данных (не всегда релевантных для нашей цели), пытается представить эти данные в более сжатом виде, путем сохранения как можно большего количества информации.
![scikit-learn logo](pics/dimens_red.jpg)
