#### Creating my features

```python
def make_features(df):
    #df['missing_flag'] = df.isnull().sum(axis=1)
    df['sex_life'] = df['Age'] - df['First sexual intercourse']
    df['total_packs'] = df['Smokes (years)'] * df['Smokes (packs/year)']
    df['pregnancies_per_year'] = df['Num of pregnancies'] / df['sex_life']
    df['pregnancies_per_year'] = df['pregnancies_per_year'].fillna(0)
    df['pregnancies_per_year_2'] = df['Num of pregnancies'] / df['Age']
    df['pregnancies_per_partner'] = df['Num of pregnancies'] / df['Number of sexual partners']
    
    
    df['STD_per_year'] = df['STDs (number)'] / df['sex_life']
    df['STD_per_year'] = df['STD_per_year'].fillna(0)
    df['STD_per_year_2'] = df['STDs (number)'] / df['Age']
    df['STD_per_year_3'] = df['STDs (number)'] / \
        (df['STDs: Time since first diagnosis'] - df['STDs: Time since last diagnosis'])
    df['STD_per_year_3'] = df['STD_per_year_3'].fillna(0) 
    df['partners_per_year'] = df['Number of sexual partners'] / df['sex_life']
    df['partners_per_year'] = df['partners_per_year'].fillna(0)
    df['partners_per_year_2'] = df['Number of sexual partners'] / df['Age']
    df['started_moking_age'] = df['Age'] - df['Smokes (years)']
    df['started_hc_age'] = df['Age'] - df['Hormonal Contraceptives (years)']
    df = df.replace(np.inf, 0)
    return df
```

```python
data = make_features(data)
```

## Model creation

I will estimate F1 metrics since we have class imbalance. Also 1 is more important than 0 since we have medical problem: its better to diagnose false cancer that miss true cancer

Trying CatBoost classyfier

```python
import catboost
from sklearn import model_selection
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import neighbors
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[features], data['my_target'], test_size=0.33, random_state=1234)
```

Will use GridSearchCV to perform both CV and grid search over params

- Decision tree

```python
parameters = {'max_depth': [int(x) for x in np.arange(1, 11)]}

clf = GridSearchCV(DecisionTreeClassifier(random_state=1234), parameters, cv=5, verbose=5, n_jobs=3, scoring='f1')
clf.fit(X_train, y_train.ravel())
print('best params {}'.format(clf.best_params_))
```

```python
clf_tree = DecisionTreeClassifier(max_depth=8)
clf_tree.fit(X_train, y_train)
predicted_train_tree = clf_tree.predict(X_train)
predicted_test_tree = clf_tree.predict(X_test)
print('F1 for DecisionTree for both classes train: {}, {}'.format(
metrics.precision_recall_fscore_support(y_train, predicted_train_tree)[2][0],
metrics.precision_recall_fscore_support(y_train, predicted_train_tree)[2][1] 
))
print('F1 for DecisionTree for both classes test: {}, {}'.format(
metrics.precision_recall_fscore_support(y_test, predicted_test_tree)[2][0],
metrics.precision_recall_fscore_support(y_test, predicted_test_tree)[2][1] 
))
```

- XGBoosting

```python
parameters = {'max_depth': [int(x) for x in np.arange(1, 11)], 
             'learning_rate': [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3],
             'objective': ['binary:logistic', 'binary:logitraw'],
              'min_child_weight':[1,5,100]
             }

clf2 = GridSearchCV(XGBClassifier(random_state=1234), parameters, cv=5, verbose=5, n_jobs=3, scoring='f1')
clf2.fit(X_train, y_train.ravel())
print('best params {}'.format(clf2.best_params_))
```

```python
clf_xgb = XGBClassifier(objective='binary:logistic', learning_rate=0.0001, max_depth=7, random_state=1234)
clf_xgb.fit(X_train, y_train)
predicted_train_xgb = clf_xgb.predict(X_train)
predicted_test_xgb = clf_xgb.predict(X_test)
print('F1 for XGB for both classes train: {}, {}'.format(
metrics.precision_recall_fscore_support(y_train, predicted_train_xgb)[2][0],
metrics.precision_recall_fscore_support(y_train, predicted_train_xgb)[2][1] 
))
print('F1 for XGB for both classes test: {}, {}'.format(
metrics.precision_recall_fscore_support(y_test, predicted_test_xgb)[2][0],
metrics.precision_recall_fscore_support(y_test, predicted_test_xgb)[2][1] 
))
```

- kNN

```python
parameters = {'n_neighbors': [int(x) for x in np.arange(1, 11)], 
             'weights': ['uniform', 'distance'],
              'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']
             }

clf7 = GridSearchCV(neighbors.KNeighborsClassifier(), parameters, cv=5, verbose=5, n_jobs=3, scoring='f1')
clf7.fit(X_train, y_train.ravel())
print('best params {}'.format(clf7.best_params_))
```

```python
clf_knn = neighbors.KNeighborsClassifier(weights='uniform', n_neighbors=1, algorithm='auto')
clf_knn.fit(X_train, y_train)
predicted_train_knn = clf_knn.predict(X_train)
predicted_test_knn = clf_knn.predict(X_test)
print('F1 for XGB for both classes train: {}, {}'.format(
metrics.precision_recall_fscore_support(y_train, predicted_train_knn)[2][0],
metrics.precision_recall_fscore_support(y_train, predicted_train_knn)[2][1] 
))
print('F1 for XGB for both classes test: {}, {}'.format(
metrics.precision_recall_fscore_support(y_test, predicted_test_knn)[2][0],
metrics.precision_recall_fscore_support(y_test, predicted_test_knn)[2][1] 
))
```

Trying to oversample SMOT

```python
from imblearn.over_sampling import SMOTE

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test #= train_test_split(data[features], data[, test_size=0.3, random_state=0)
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel())
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

### Conclusion

As we mentioned it is more important to focuse on 1 rather than on 0. Classes are imbalanced (about 11% of 1). 
Methods like oversampling can be used. Best algorithm is DecisionTree learned on oversampled data. 

In fields like medicine it is important to minimize FalseNegative errors. 
So this model can be used to make more accurate prediction over cancer diagnostics because as we saw from initial dataset classic methods can be very inaccurate.
